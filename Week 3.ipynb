```python
clothing = {"hat" : "yellow", "socks" : "pink", "pants": "brown", "shirt":"blue", "jacket": "black"}

colors = ( "orange", "green", "purple", "lilac", "pink", "red")
```

```python
#print out each item with its color
#make sure it's gramatically correct

for item in clothing: 
    #check if ploral 
    if item[-1] is "s":
        print "The " + item + " are " + clothing[item] + "."
    else: 
        print "The " + item + " is " + clothing[item] + "."
        
```

```python
# loop through each item and store off verb 
for item in clothing:
    if item[-1] is "s":
        verb = " are "
    else:
        verb = " is "
    print "The " + item + verb + clothing[item] + "."
```

```python
# only print out clothing items that are in colors list
for item in clothing: 
    #check if in color list 
    if clothing[item] in colors:
        print item
        
```

```python
#put items of clothing that are in the colors list into a sentence
goodclothes = []

for item in clothing:
    # check if in color list
    if clothing[item] in colors:
        goodclothes.append(item)
        
print goodclothes
print "The best colored clothes are my " + " and ".join(goodclothes) + "."
```
